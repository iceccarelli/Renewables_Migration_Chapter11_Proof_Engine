# 02_Sun_Migration_Analysis.ipynb: Entropy, Carbon, and Stability Deep Dives

This notebook provides a deeper analytical dive into the concepts of entropy reduction, carbon command, and stability thresholds as presented in **Chapter 11: The Migration of the Sun** of "The Renewables Migration" by Vincenzo Grimaldi. It complements `01_Prove_Chapter11.ipynb` by offering more advanced simulations and visualizations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys

# Add the parent directory to the system path to import chapter11_core
script_dir = os.path.dirname(__file__)
sys.path.insert(0, os.path.abspath(os.path.join(script_dir, "..")))
from chapter11_core import Chapter11Core

# Initialize core logic
core = Chapter11Core()

# Configure matplotlib for better aesthetics
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 7)
plt.rcParams["figure.dpi"] = 100
```

## Entropy Reduction Dynamics (Maxwell's Demon Model)

```python
# Simulate entropy reduction over varying protocol depths
initial_entropy = np.linspace(0.5, 1.0, 5) # Different starting entropy levels
protocol_depths = np.linspace(0, 1, 100) # Varying protocol depth factors

plt.figure()
for ent in initial_entropy:
    reduced_entropies = [core.entropy_reduction_simulation(ent, pd) for pd in protocol_depths]
    sns.lineplot(x=protocol_depths, y=reduced_entropies, label=f"Initial Entropy: {ent:.1f}")

plt.xlabel("Protocol Depth Factor")
plt.ylabel("Reduced Entropy Level")
plt.title("Entropy Reduction Dynamics with Varying Protocol Depth")
plt.legend()
plt.savefig("../plots/final_entropy_old_model.png")
plt.show()
print("Generated final_entropy_old_model.png")
```

## Carbon Command Equation Analysis

```python
# Analyze Carbon Command under different scenarios
delta_co2_domestic_scenarios = [-200, -240, -280] # Mt
delta_co2_embodied_scenarios = [30, 45, 60] # Mt
phi_mcp_range = np.linspace(0, 2, 100)

plt.figure()
for dom_co2, emb_co2 in zip(delta_co2_domestic_scenarios, delta_co2_embodied_scenarios):
    carbon_commands = [core.carbon_command_equation(dom_co2, emb_co2, pm) for pm in phi_mcp_range]
    sns.lineplot(x=phi_mcp_range, y=carbon_commands, label=f"Domestic: {dom_co2}Mt, Embodied: {emb_co2}Mt")

plt.xlabel("ΦMCP (Transparency Multiplier)")
plt.ylabel("Carbon Command (Ccmd) (Mt)")
plt.title("Carbon Command Equation under Various Scenarios")
plt.legend()
plt.savefig("../plots/carbon_leakage_recovery.png") # Reusing plot name from Chapter 8 for consistency if applicable
plt.show()
print("Generated carbon_leakage_recovery.png")
```

## Stability Threshold Projection (Multi-variable Analysis)

```python
# Explore stability across varying complexity and protocol factors
psi_mcp_levels = np.linspace(1.0, 2.0, 5) # Different Protocol Stability Factor levels
complexities = np.linspace(0.1, 1.0, 100) # Varying complexity

plt.figure()
for psi in psi_mcp_levels:
    stabilities = [core.stability_threshold_projection(0.9, 0.8, 0.7, psi, comp, 0.6) for comp in complexities]
    sns.lineplot(x=complexities, y=stabilities, label=f"ΨMCP: {psi:.1f}")

plt.xlabel("Complexity")
plt.ylabel("Stability Threshold (S(t))")
plt.title("Stability Threshold Projection with Varying Complexity and ΨMCP")
plt.legend()
plt.savefig("../plots/stability_threshold_projection.png")
plt.show()
print("Generated stability_threshold_projection.png")
```
