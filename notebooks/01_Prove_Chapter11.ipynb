## Isolated Sprint vs Protocol Leadership Curve (Figure 11.2)

```python
# Interactive sliders for simulation
protocol_unlock_slider = st.slider("Protocol Unlock Level (0-100%)", 0.0, 1.0, core.protocol_unlock_2030, 0.01)
leadership_factor_slider = st.slider("Leadership Factor (0-5)", 0.0, 5.0, core.leadership_factor_base, 0.1)

# Calculate performance
performance = core.isolated_sprint_vs_protocol_leadership(protocol_unlock_slider, leadership_factor_slider)
st.write(f"**Simulated Performance Index:** {performance:.2f}")

# Plotting the curve (conceptual)
protocol_unlock_range = np.linspace(0, 1, 100)
leadership_factor_fixed = core.leadership_factor_base

performance_curve = [core.isolated_sprint_vs_protocol_leadership(pu, leadership_factor_fixed) for pu in protocol_unlock_range]

plt.figure(figsize=(10, 6))
sns.lineplot(x=protocol_unlock_range, y=performance_curve, label="Protocol Leadership Performance")
plt.xlabel("Protocol Unlock Level")
plt.ylabel("Performance Index")
plt.title("Isolated Sprint vs Protocol Leadership (Conceptual)")
plt.grid(True)
plt.legend()
st.pyplot(plt)
```

## Entropy and Maxwell’s Demon Simulation

```python
entropy_level_slider = st.slider("Initial Entropy Level (0-1)", 0.0, 1.0, core.entropy_level_base, 0.01)
protocol_depth_factor_slider = st.slider("Protocol Depth Factor (0-1)", 0.0, 1.0, 0.5, 0.01)

reduced_entropy = core.entropy_reduction_simulation(entropy_level_slider, protocol_depth_factor_slider)
st.write(f"**Reduced Entropy Level:** {reduced_entropy:.2f}")

# Plotting entropy reduction
protocol_depth_range = np.linspace(0, 1, 100)
entropy_reduction_curve = [core.entropy_reduction_simulation(entropy_level_slider, pd) for pd in protocol_depth_range]

plt.figure(figsize=(10, 6))
sns.lineplot(x=protocol_depth_range, y=entropy_reduction_curve, label="Entropy Reduction")
plt.xlabel("Protocol Depth Factor")
plt.ylabel("Entropy Level")
plt.title("Entropy Reduction via Protocol (Maxwell's Demon Model)")
plt.grid(True)
plt.legend()
st.pyplot(plt)
```

## Carbon Command Equation Prover

```python
delta_co2_domestic_input = st.number_input("∆CO2,domestic (Mt)", value=-240.0, step=1.0)
delta_co2_embodied_input = st.number_input("∆CO2,embodied (Mt)", value=45.0, step=1.0)
phi_mcp_input = st.slider("ΦMCP (Transparency Multiplier)", 0.0, 2.0, core.phi_mcp_2030, 0.01)

ccmd_calculated = core.carbon_command_equation(delta_co2_domestic_input, delta_co2_embodied_input, phi_mcp_input)
st.write(f"**Carbon Command (Ccmd) (Mt):** {ccmd_calculated:.2f}")

# Plotting Carbon Command vs. ΦMCP
phi_mcp_range = np.linspace(0, 2, 100)
carbon_command_curve = [core.carbon_command_equation(delta_co2_domestic_input, delta_co2_embodied_input, pm) for pm in phi_mcp_range]

plt.figure(figsize=(10, 6))
sns.lineplot(x=phi_mcp_range, y=carbon_command_curve, label="Carbon Command")
plt.xlabel("ΦMCP (Transparency Multiplier)")
plt.ylabel("Carbon Command (Mt)")
plt.title("Protocol-Enabled Carbon Command Equation")
plt.grid(True)
plt.legend()
st.pyplot(plt)
```

## Stability Threshold Projection

```python
efficiency_eta_input = st.slider("ηeff (Efficiency)", 0.0, 1.0, 0.9, 0.01)
coupling_factor_input = st.slider("Fcoupling (Coupling Factor)", 0.0, 1.0, 0.8, 0.01)
storage_stability_input = st.slider("Sstorage (Storage Stability)", 0.0, 1.0, 0.7, 0.01)
psi_mcp_input = st.slider("ΨMCP (Protocol Stability Factor)", 0.0, 2.0, core.psi_mcp_2030, 0.01)
complexity_input = st.slider("Complexity". 0.1, 1.0, 0.5, 0.01)
dependency_input = st.slider("Dependency". 0.1, 1.0, 0.6, 0.01)

stability_calculated = core.stability_threshold_projection(efficiency_eta_input, coupling_factor_input, storage_stability_input, psi_mcp_input, complexity_input, dependency_input)
st.write(f"**Stability Threshold (S(t)):** {stability_calculated:.2f}")

# Plotting Stability Threshold vs. ΨMCP
psi_mcp_range = np.linspace(0, 2, 100)
stability_curve = [core.stability_threshold_projection(efficiency_eta_input, coupling_factor_input, storage_stability_input, pm, complexity_input, dependency_input) for pm in psi_mcp_range]

plt.figure(figsize=(10, 6))
sns.lineplot(x=psi_mcp_range, y=stability_curve, label="Stability Threshold")
plt.xlabel("ΨMCP (Protocol Stability Factor)")
plt.ylabel("Stability Threshold (S(t))")
plt.title("Stability Threshold Projection")
plt.grid(True)
plt.legend()
st.pyplot(plt)
```
